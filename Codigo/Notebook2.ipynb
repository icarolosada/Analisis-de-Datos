```python
import numpy as np
import matplotlib.pyplot as plt

# Ordenar el DataFrame por años en orden ascendente
media_lluvia_anos_df = media_lluvia_anos_df.sort_values('años')

# Obtener los valores de los años y la media de lluvia
años = media_lluvia_anos_df['años']
media_lluvia = media_lluvia_anos_df['media_lluvia']

# Calcular los coeficientes de la línea de tendencia
coeficientes = np.polyfit(años, media_lluvia, 1)
linea_tendencia = np.polyval(coeficientes, años)

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Graficar la media de lluvia como un diagrama de líneas
plt.plot(años, media_lluvia, label='Media de lluvia')
plt.plot(años, linea_tendencia, color='red', label='Línea de tendencia')

# Configurar etiquetas y título del gráfico
plt.xlabel('Años')
plt.ylabel('Media de lluvia')
plt.title('Media de lluvia por años')

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ordenar el DataFrame por años en orden ascendente
media_lluvia_anos_df = media_lluvia_anos_df.sort_values('años')

# Obtener los valores de los años y la media de lluvia
años = media_lluvia_anos_df['años']
media_lluvia = media_lluvia_anos_df['media_lluvia']

# Calcular el suavizado exponencial ponderado
alpha = 0.1  # Factor de ponderación
media_suavizada = media_lluvia.ewm(alpha=alpha, adjust=False).mean()

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Graficar la media de lluvia como un diagrama de líneas
plt.plot(años, media_lluvia, label='Media de lluvia')
plt.plot(años, media_suavizada, color='red', label='Línea de tendencia suavizada')

# Configurar etiquetas y título del gráfico
plt.xlabel('Años')
plt.ylabel('Media de lluvia')
plt.title('Media de lluvia por años')

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Obtener los valores de los años, media de temperatura máxima y media de temperatura mínima
años = Media_temperaturas_df['año']
media_temp_max = Media_temperaturas_df['media_temp_max']
media_temp_min = Media_temperaturas_df['media_temp_min']
temp_media = Media_temperaturas_df['temp_media']

# Calcular la media móvil utilizando un promedio de 3 puntos adyacentes
temp_media_suavizada = temp_media.rolling(window=15, center=True).mean()

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Graficar la línea de tendencia de la media de temperatura máxima
plt.plot(años, media_temp_max, color='red', label='Línea de tendencia - Temp. máxima')

# Graficar la línea de tendencia de la media de temperatura mínima
plt.plot(años, media_temp_min, color='blue', label='Línea de tendencia - Temp. mínima')

# Graficar la línea de tendencia suavizada de temp_media
plt.plot(años, temp_media_suavizada, color='black', label='Línea de tendencia suavizada - Temp. media')

# Configurar etiquetas y título del gráfico
plt.xlabel('Año')
plt.ylabel('Media de temperatura')
plt.title('Media de temperatura máxima, mínima y media por año')

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Obtener los valores de los años y el agua total
años = Capacida_embalses_df['año']
agua_total = Capacida_embalses_df['total_agua_total']

# Calcular la media y la desviación estándar del agua total
media = agua_total.mean()
desviacion = agua_total.std()

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Graficar la evolución del agua total
plt.plot(años, agua_total, color='blue')

# Calcular la línea de tendencia como la media +/- una desviación de varianza
linea_tendencia = media + desviacion * np.linspace(-1, 1, len(años))

# Graficar la línea de tendencia en color rojo
plt.plot(años, linea_tendencia, color='red', label='Línea de tendencia')

# Configurar etiquetas y título del gráfico
plt.xlabel('Año')
plt.ylabel('Agua total')
plt.title('Evolución del agua total en embalses')

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```
