```python
import numpy as np
import matplotlib.pyplot as plt

# Ordenar el DataFrame por años en orden ascendente
media_lluvia_anos_df = media_lluvia_anos_df.sort_values('años')

# Obtener los valores de los años y la media de lluvia
años = media_lluvia_anos_df['años']
media_lluvia = media_lluvia_anos_df['media_lluvia']

# Calcular los coeficientes de la línea de tendencia
coeficientes = np.polyfit(años, media_lluvia, 1)
linea_tendencia = np.polyval(coeficientes, años)

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Graficar la media de lluvia como un diagrama de líneas
plt.plot(años, media_lluvia, label='Media de lluvia')
plt.plot(años, linea_tendencia, color='red', label='Línea de tendencia')

# Configurar etiquetas y título del gráfico
plt.xlabel('Años')
plt.ylabel('Media de lluvia 10l/m2')
plt.title('Media de lluvia por años tendecia')

# Mostrar la leyenda
plt.legend()

# Guardar el gráfico en un archivo de imagen
plt.savefig('C:\\Users\\Icaro.DESKTOP-8VQB1F3\\Desktop\\Data Analyst\\Histórico del Clima\\Github\\Visualizaciones\\Visualizaciones\\Media_lluvia_años_tendencia.png')

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ordenar el DataFrame por años en orden ascendente
media_lluvia_anos_df = media_lluvia_anos_df.sort_values('años')

# Obtener los valores de los años y la media de lluvia
años = media_lluvia_anos_df['años']
media_lluvia = media_lluvia_anos_df['media_lluvia']

# Calcular el suavizado exponencial ponderado
alpha = 0.1  # Factor de ponderación
media_suavizada = media_lluvia.ewm(alpha=alpha, adjust=False).mean()

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Graficar la media de lluvia como un diagrama de líneas
plt.plot(años, media_lluvia, label='Media de lluvia')
plt.plot(años, media_suavizada, color='red', label='Línea de tendencia suavizada')

# Configurar etiquetas y título del gráfico
plt.xlabel('Años')
plt.ylabel('Media de lluvia 10l/m2')
plt.title('Media de lluvia por años y promedio')

# Mostrar la leyenda
plt.legend()


# Guardar el gráfico en un archivo de imagen
plt.savefig('C:\\Users\\Icaro.DESKTOP-8VQB1F3\\Desktop\\Data Analyst\\Histórico del Clima\\Github\\Visualizaciones\\Visualizaciones\\Media_lluvia_años_promedio.png')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Obtener los valores de los años y el agua total
años = Capacida_embalses_df['Año']
agua_total = Capacida_embalses_df['Agua embalsada']

# Calcular la media y la desviación estándar del agua total
media = agua_total.mean()
desviacion = agua_total.std()

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Graficar la evolución del agua total
plt.plot(años, agua_total, color='blue')

# Calcular la línea de tendencia como la media +/- una desviación de varianza
linea_tendencia = media + desviacion * np.linspace(-1, 1, len(años))

# Graficar la línea de tendencia en color rojo
plt.plot(años, linea_tendencia, color='red', label='Línea de tendencia')

# Configurar etiquetas y título del gráfico
plt.xlabel('Año')
plt.ylabel('Agua Hm3')
plt.title('Capacidad embalses')

# Mostrar la leyenda
plt.legend()


# Guardar el gráfico en un archivo de imagen
plt.savefig('C:\\Users\\Icaro.DESKTOP-8VQB1F3\\Desktop\\Data Analyst\\Histórico del Clima\\Github\\Visualizaciones\\Visualizaciones\\capacidad_embalses.png', format='png')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Obtener los valores de los años, semanas y agua embalsada
años = Agua_embalsada_df['Año']
semanas = Agua_embalsada_df['Semana']
agua_embalsada = Agua_embalsada_df['Agua embalsada']

# Agrupar los datos por año y semana y calcular la media del agua embalsada
datos_agrupados = Agua_embalsada_df.groupby(['Año', 'Semana'])['Agua embalsada'].mean()

# Restablecer el índice para tener los años y semanas como columnas
datos_pivot = datos_agrupados.reset_index().pivot(index='Semana', columns='Año', values='Agua embalsada')

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 8))

# Graficar las líneas para cada año
for año in datos_pivot.columns:
    plt.plot(datos_pivot.index, datos_pivot[año], label=str(año))

# Configurar etiquetas y título del gráfico
plt.xlabel('Semana')
plt.ylabel('Agua Hm3')
plt.title('Evolución del agua embalsada por año')

# Establecer etiquetas personalizadas en el eje x con espaciado
espaciado = 5  # Espacio entre las etiquetas
plt.xticks(datos_pivot.index[::espaciado], [f'Semana {semana}' for semana in datos_pivot.index[::espaciado]], rotation=45)

# Mostrar la leyenda fuera del área del gráfico
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar automáticamente los espacios en el gráfico
plt.tight_layout()

# Guardar el gráfico en un archivo de imagen
plt.savefig('C:\\Users\\Icaro.DESKTOP-8VQB1F3\\Desktop\\Data Analyst\\Histórico del Clima\\Github\\Visualizaciones\\Visualizaciones\\Evolución_agua_embalsada.png', format='png')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Obtener los valores de los años, semanas y agua embalsada
años = Agua_embalsada_df['Año']
semanas = Agua_embalsada_df['Semana']
agua_embalsada = Agua_embalsada_df['Agua embalsada']

# Agrupar los datos por año y semana y calcular la media del agua embalsada
datos_agrupados = Agua_embalsada_df.groupby(['Año', 'Semana'])['Agua embalsada'].mean()

# Restablecer el índice para tener los años y semanas como columnas
datos_pivot = datos_agrupados.reset_index().pivot(index='Semana', columns='Año', values='Agua embalsada')

# Agrupar los años en períodos de 5 años
datos_pivot_grouped = datos_pivot.groupby((datos_pivot.columns // 5) * 5, axis=1).mean()

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Graficar las líneas para cada grupo de 5 años
for grupo in datos_pivot_grouped.columns:
    plt.plot(datos_pivot.index, datos_pivot_grouped[grupo], label=f'{grupo}-{grupo+4}')

# Configurar etiquetas y título del gráfico
plt.xlabel('Semana')
plt.ylabel('Agua Hm3')
plt.title('Evolución del agua embalsada por período de 5 años')

# Establecer etiquetas personalizadas en el eje x con espaciado
espaciado = 5  # Espacio entre las etiquetas
plt.xticks(datos_pivot.index[::espaciado], [f'Semana {semana}' for semana in datos_pivot.index[::espaciado]], rotation=45)

# Mostrar la leyenda fuera del área del gráfico
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar automáticamente los espacios en el gráfico
plt.tight_layout()


# Guardar el gráfico en un archivo de imagen
plt.savefig('C:\\Users\\Icaro.DESKTOP-8VQB1F3\\Desktop\\Data Analyst\\Histórico del Clima\\Github\\Visualizaciones\\Visualizaciones\\Evolución_agua_embalsada_5años.png', format='png')

# Mostrar el gráfico
plt.show()
```
